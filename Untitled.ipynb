```python
import matplotlib.pyplot as plt
```

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600,  400, 400,700]
Rohit = [0,0,400,700,1000,1500,1400,1700,1800,2000]
years = [ 2005, 2007, 2010,2011,2012,2013,2014,2015,2016,2017]

plt.plot(years, kohli, color='orange',linewidth=5, linestyle='--' ,label="Virat kholi")
plt.plot(years, sehwag, color='yellow',linewidth=3, linestyle='-.' ,label="Virender sehwag")
plt.plot(years, Rohit, color='pink', label="Rohit sharma")


plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.legend()
plt.style.use("seaborn-v0_8")
plt.grid(True)
plt.show()
```
